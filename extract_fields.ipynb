# Extract all Velocity Field Data

Extract all velocity field data from Incompact3D .bin files

```python
# Imports
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
# Enter the mesh size parameters
nx = 128
ny = 65
nz = 64
num_t = 6

x_len = 8.
y_len = 2.
z_len = 4.
```

```python
# Initialize velocity data matrices
ux = np.empty((nx,ny,nz,num_t))
uy = np.empty((nx,ny,nz,num_t))
uz = np.empty((nx,ny,nz,num_t))
```

```python
# Extract data from files
for i in range(0,num_t):
    idx = i+1
    ux_read = np.fromfile(f"../Incompact3d/examples/Channel-Flow/data/ux-00{idx}.bin")
    uy_read = np.fromfile(f"../Incompact3d/examples/Channel-Flow/data/uy-00{idx}.bin")
    uz_read = np.fromfile(f"../Incompact3d/examples/Channel-Flow/data/uz-00{idx}.bin")


    ux_reshape = ux_read.reshape(nz,ny,nx).transpose()
    uy_reshape = uy_read.reshape(nz,ny,nx).transpose()
    uz_reshape = uz_read.reshape(nz,ny,nx).transpose()

    ux[:,:,:,i] = ux_reshape
    uy[:,:,:,i] = uy_reshape
    uz[:,:,:,i] = uz_reshape
```

# Plot data to verify

```python
ux.shape
```

```python
# Create XY grid
x = np.linspace(0,x_len,nx)
y = np.linspace(0,y_len,ny)
z = np.linspace(0,z_len,nz)
X,Y,Z = np.meshgrid(x,y,z,indexing='ij')
```

```python
fig = plt.figure(figsize=(16,4))
plt.pcolormesh(X[:,:,30],Y[:,:,30],ux[:,:,32,3], cmap='coolwarm', shading='gouraud')
plt.colorbar(label='Velocity [m/s]')
plt.xlabel('X [m]')
plt.ylabel('Y [m]')
```

```python
fig = plt.figure(figsize=(15,8))
plt.imshow(plt.imread('pics/ux_XY_plane.PNG'))
```
